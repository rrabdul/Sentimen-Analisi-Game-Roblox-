2. Importing Library

```python
from google_play_scraper import Sort, reviews_all
import csv
import pandas as pd
```

3. Scrapping Details

```python
# Mengambil semua ulasan dari aplikasi Roblox
scrapreview = reviews_all(
    'com.roblox.client',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=150000
)
scrapreview_limited = scrapreview[:150000]
```

4. Output Details

```python
dataView = pd.DataFrame(scrapreview_limited)
dataView.head()
```

```python
with open('hasil_scraping_Roblox.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)

    writer.writerow(['content', 'score'])

    for review in scrapreview_limited:
        content = review.get('content')
        if content and 'roblox' in content.lower():  # Cek jika konten tidak None dan mengandung 'roblox'
            writer.writerow([
                content,
                review.get('score', '')
            ])

print("Data yang mengandung 'Roblox' telah disimpan ke hasil_scraping_Roblox.csv")
```
